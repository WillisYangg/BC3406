#### log reg function

```python
health_factors = ['ADI', 'Age', 'Gender', 'lsbio_cholesterol', 'lsbio_glucose', 'lsbio_bp', 'lsbio_weight', 'lsbio_trig', 
                  'lshrq_physact', 'lshrq_tobacco', 'lshrq_nutrition', 'lshrq_stress', 'lshrq_genHealth', 'lshrq_wellbeing', 'lshrq_alcohol', 'lshrq_sleeping', 
                  'cc_RespiratorySinusAllergy_chronic', 'cc_diabetes', 'cc_cancer', 'cc_mskChronic', 'cc_cholesterol', 'cc_cardiac', 'cc_depression', 'cc_obesity', 
                  'cc_mskAllOther', 'cc_behavioralAllOther', 'cc_ckd', 'concurrent_risk', 'rx_nonClaimant', 'rx_nonCompliant', 'extreme_heat_2018']
```

```python
def log_model(dataframe, y_column, data_list):
    print(f'Y variable: {y_column}')
    print('---------------------------------------------------------------------------------')

    y = pd.DataFrame(dataframe[y_column])
    health_environment_factors = [col for col in health_factors if col not in data_list]
    x = pd.DataFrame(dataframe[health_environment_factors])
    x[health_environment_factors] = x[health_environment_factors].astype('category')
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
    logmodel = LogisticRegression()
    logmodel.fit(x_train, y_train)
    print('Intercept of Regression \t: b = ', logmodel.intercept_)

    x_with_constant = sm.add_constant(x_train)
    model = sm.Logit(y_train, x_with_constant)
    result = model.fit()

    significant_variable_lr = {}

    p_value_dataframe = pd.DataFrame(result.pvalues)
    p_value_dataframe = p_value_dataframe.rename(columns={0: 'p-value'})
    p_value_dataframe = p_value_dataframe.round(2)

    significant_variable_lr['significant_variable'] = p_value_dataframe[(p_value_dataframe['p-value'] < 0.1)].index.tolist()

    print(significant_variable_lr)
    print('---------------------------------------------------------------------------------')
```

```python
for health_variables in significant_health_variables_list:
    try:
        log_model(df_P, health_variables, significant_health_variables_list)
    except Exception as e:
        print(e)
```

```python
for health_variables in significant_health_variables_list_indoor:
    try:
        log_model(df_C, health_variables, significant_health_variables_list_indoor)
    except Exception as e:
        print(e)
```
