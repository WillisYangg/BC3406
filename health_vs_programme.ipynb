```python
df = df.sort_values(by=['person_id', 'analytic_period_id'])

grouped = df.groupby('person_id')

health_factors = ['lsbio_cholesterol', 'lsbio_glucose', 'lsbio_bp', 'lsbio_weight', 'lsbio_trig', 'lshrq_physact', 'lshrq_tobacco', 'lshrq_nutrition', 
                  'lshrq_stress', 'lshrq_genHealth', 'lshrq_wellbeing', 'lshrq_alcohol', 'lshrq_sleeping', 'cc_RespiratorySinusAllergy_chronic', 
                  'cc_diabetes', 'cc_cancer', 'cc_mskChronic', 'cc_cholesterol', 'cc_cardiac', 'cc_depression', 'cc_obesity', 
                  'cc_mskAllOther', 'cc_behavioralAllOther', 'cc_ckd', 'concurrent_risk', 'rx_nonClaimant', 'rx_nonCompliant']

for person_id, group_df in grouped:
    for health in health_factors:
        if (group_df['analytic_period_id'] == 1).any() and (group_df.loc[group_df['analytic_period_id'] == 1, health] == 1).any():
            df.loc[(df['person_id'] == person_id) & (df['analytic_period_id'] != 1), health] = 1
        if (group_df['analytic_period_id'] == 2).any() and (group_df.loc[group_df['analytic_period_id'] == 2, health] == 1).any():
            df.loc[(df['person_id'] == person_id) & (df['analytic_period_id'] == 3), health] = 1
```

### Subgroup P

#### cc_mskChronic

```python
df_P_mskChronic = df_P.loc[(df_P['cc_mskChronic'] == 1) & 
                           (df_P['cc_diabetes'] == 0) & 
                           (df_P['cc_cardiac'] == 0)].reset_index(drop=True)

df_P_mskChronic = df_P_mskChronic.sort_values(by=['person_id', 'analytic_period_id'])

df_P_mskChronic['percentage_diff_absence'] = round((df_P_mskChronic.groupby('person_id')['absence_indicator'].diff() * -1),0)
df_P_mskChronic['percentage_diff_absence'] = df_P_mskChronic['percentage_diff_absence'].fillna(0)
df_P_mskChronic['percentage_diff_absence'] = np.where(df_P_mskChronic['percentage_diff_absence']== -0, 0, df_P_mskChronic['percentage_diff_absence'])
df_P_mskChronic[health_programs] = np.where(df_P_mskChronic[health_programs] == -1, 0, df_P_mskChronic[health_programs])

df_P_mskChronic
```

```python
counts = df_P_mskChronic.groupby('person_id')['analytic_period_id'].unique()
person_ids_with_both_periods = [pid for pid, periods in counts.items() if 1 in periods and 2 in periods]

df_P_mskChronic_1_2 = df_P_mskChronic[df_P_mskChronic['person_id'].isin(person_ids_with_both_periods)]
df_P_mskChronic_1_2.reset_index(drop=True, inplace=True)
df_P_mskChronic_1_2 = df_P_mskChronic_1_2.loc[(df_P_mskChronic_1_2['analytic_period_id']==1) | (df_P_mskChronic_1_2['analytic_period_id']==2)]

df_P_mskChronic_1_2
```

```python
health_comparison_factors = health_programs + ['person_id']

year_1 = df_P_mskChronic_1_2.loc[df_P_mskChronic_1_2['analytic_period_id']==1][health_comparison_factors].reset_index(drop=True)
year_2 = df_P_mskChronic_1_2.loc[df_P_mskChronic_1_2['analytic_period_id']==2]['percentage_diff_absence'].reset_index(drop=True)

df_1_2 = year_1.merge(year_2, left_index=True, right_index=True,
                 how='outer', suffixes=('', '_y'))

df_1_2.drop(df_1_2.filter(regex='_y$').columns, axis=1, inplace=True)

df_1_2
```

```python
counts = df_P_mskChronic.groupby('person_id')['analytic_period_id'].unique()
person_ids_with_both_periods = [pid for pid, periods in counts.items() if 2 in periods and 3 in periods]

df_P_mskChronic_2_3 = df_P_mskChronic[df_P_mskChronic['person_id'].isin(person_ids_with_both_periods)]
df_P_mskChronic_2_3.reset_index(drop=True, inplace=True)
df_P_mskChronic_2_3 = df_P_mskChronic_2_3.loc[(df_P_mskChronic_2_3['analytic_period_id']==2) | (df_P_mskChronic_2_3['analytic_period_id']==3)]

df_P_mskChronic_2_3
```

```python
year_2 = df_P_mskChronic_2_3.loc[df_P_mskChronic_2_3['analytic_period_id']==2][health_comparison_factors].reset_index(drop=True)
year_3 = df_P_mskChronic_2_3.loc[df_P_mskChronic_2_3['analytic_period_id']==3]['percentage_diff_absence'].reset_index(drop=True)

df_2_3 = year_2.merge(year_3, left_index=True, right_index=True,
                 how='outer', suffixes=('', '_y'))

df_2_3.drop(df_2_3.filter(regex='_y$').columns, axis=1, inplace=True)

df_2_3
```

```python
# train data: 2019 - 2018
# test data: 2020 - 2019

# get data from person_id with both analytic_period_id 1 and 2

from sklearn.model_selection import train_test_split

y_train = pd.DataFrame(df_1_2['percentage_diff_absence'])
y_train.loc[y_train['percentage_diff_absence'] > 0, 'percentage_diff_absence'] = np.log(y_train.loc[y_train['percentage_diff_absence'] > 0, 'percentage_diff_absence'])
y_train.loc[y_train['percentage_diff_absence'] < 0, 'percentage_diff_absence'] = -np.log(-y_train.loc[y_train['percentage_diff_absence'] < 0, 'percentage_diff_absence'])
x_train = pd.DataFrame(df_1_2[health_programs])
x_train[health_programs] = x_train[health_programs].astype('category')

y_test = pd.DataFrame(df_2_3['percentage_diff_absence'])
y_test.loc[y_test['percentage_diff_absence'] > 0, 'percentage_diff_absence'] = np.log(y_test.loc[y_test['percentage_diff_absence'] > 0, 'percentage_diff_absence'])
y_test.loc[y_test['percentage_diff_absence'] < 0, 'percentage_diff_absence'] = -np.log(-y_test.loc[y_test['percentage_diff_absence'] < 0, 'percentage_diff_absence'])
x_test = pd.DataFrame(df_2_3[health_programs])
x_test[health_programs] = x_test[health_programs].astype('category')

print("Train Set :", y_train.shape, x_train.shape)
print("Test Set  :", y_test.shape, x_test.shape)
```

```python
# use Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linreg = LinearRegression()

linreg.fit(x_train, y_train)

print('Intercept of Regression \t: b = ', linreg.intercept_)

y_train_pred = linreg.predict(x_train)
y_test_pred = linreg.predict(x_test)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print()
```

```python
import statsmodels.api as sm

x_train_with_constant = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train_with_constant)
result = model.fit()

print(result.summary())
```

```python
# use decision tree regressor

from sklearn.tree import DecisionTreeRegressor

decisiontreemodel = DecisionTreeRegressor(max_depth=3)
result = decisiontreemodel.fit(x_train, y_train)

y_train_pred = decisiontreemodel.predict(x_train)
train_mse = mean_squared_error(y_train_pred, y_train)
y_test_pred = decisiontreemodel.predict(x_test)
test_mse = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error (Train Dataset):", train_mse)
print("Mean Squared Error (Test Dataset):", test_mse)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': range(1, 11)}

decisiontreemodel = DecisionTreeRegressor()

grid_search = GridSearchCV(estimator=decisiontreemodel, param_grid=param_grid, cv=5)

grid_search.fit(x_train, y_train)
best_max_depth = grid_search.best_params_['max_depth']

print("Best max_depth:", best_max_depth)
```

```python
# use decision tree regressor

from sklearn.tree import DecisionTreeRegressor

decisiontreemodel = DecisionTreeRegressor(max_depth=1)
result = decisiontreemodel.fit(x_train, y_train)

y_train_pred = decisiontreemodel.predict(x_train)
train_mse = mean_squared_error(y_train_pred, y_train)
y_test_pred = decisiontreemodel.predict(x_test)
test_mse = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error (Train Dataset):", train_mse)
print("Mean Squared Error (Test Dataset):", test_mse)
```

```python
significant_variable = {}
significant_variable['significant_variables'] = []
fr = decisiontreemodel.feature_importances_ 
importance_dict = dict(zip(x_train.columns, fr))
for key in importance_dict:
    if importance_dict[key] > 0:
        significant_variable['significant_variables'].append(key)
        
print(significant_variable)
```

```python
from sklearn.ensemble import RandomForestRegressor

randomforest = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
randomforest.fit(x_train, y_train)

y_train_pred = randomforest.predict(x_train)
train_mse = mean_squared_error(y_train_pred, y_train)
y_test_pred = randomforest.predict(x_test)
test_mse = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error (Train Dataset):", train_mse)
print("Mean Squared Error (Test Dataset):", test_mse)
```

```python
randomforest = RandomForestRegressor(n_estimators=best_params['n_estimators'], max_depth=best_params['max_depth'], min_samples_split=best_params['min_samples_split'], random_state=42)
randomforest.fit(x_train, y_train)

y_train_pred = randomforest.predict(x_train)
train_mse = mean_squared_error(y_train_pred, y_train)
y_test_pred = randomforest.predict(x_test)
test_mse = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error (Train Dataset):", train_mse)
print("Mean Squared Error (Test Dataset):", test_mse)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': [None] + list(range(5, 20)),
    'min_samples_split': randint(2, 20)
}

randomized_search = RandomizedSearchCV(estimator=RandomForestRegressor(), param_distributions=param_dist, n_iter=5, cv=5, scoring='neg_mean_squared_error', random_state=42)

randomized_search.fit(x_train, y_train)

best_params = randomized_search.best_params_
print(best_params)
```

```python
randomforest = RandomForestRegressor(n_estimators=best_params['n_estimators'], max_depth=best_params['max_depth'], min_samples_split=best_params['min_samples_split'], random_state=42)
randomforest.fit(x_train, y_train)

y_train_pred = randomforest.predict(x_train)
train_mse = mean_squared_error(y_train_pred, y_train)
y_test_pred = randomforest.predict(x_test)
test_mse = mean_squared_error(y_test_pred, y_test)

print("Mean Squared Error (Train Dataset):", train_mse)
print("Mean Squared Error (Test Dataset):", test_mse)
```
