#### cc_mskChronic

```python
df_P_mskChronic = df_P.loc[(df_P['cc_mskChronic'] == 1) & 
                           (df_P['cc_diabetes'] != 1) & 
                           (df_P['cc_cardiac'] != 1)].reset_index(drop=True)

df_P_mskChronic = df_P_mskChronic.sort_values(by=['person_id', 'analytic_period_id'])

df_P_mskChronic['percentage_diff_absence'] = round((df_P_mskChronic.groupby('person_id')['absence_indicator'].diff() * -1),0)

df_P_mskChronic['percentage_diff_absence'] = df_P_mskChronic['percentage_diff_absence'].fillna(0)

df_P_mskChronic['percentage_diff_absence'] = np.where(df_P_mskChronic['percentage_diff_absence']== -0, 0, df_P_mskChronic['percentage_diff_absence'])

df_P_mskChronic[health_programs] = np.where(df_P_mskChronic[health_programs] == -1, 0, df_P_mskChronic[health_programs])

df_P_mskChronic
```
